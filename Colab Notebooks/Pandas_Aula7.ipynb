```python
import pandas as pd
```

```python
df1 = pd.read_excel("Aracaju.xlsx")
```

```python
df2 = pd.read_excel("Fortaleza.xlsx")
```

```python
df3 = pd.read_excel("Natal.xlsx")
df4 = pd.read_excel("Recife.xlsx")
df5 = pd.read_excel("Salvador.xlsx")
```

```python
df2.head()
```

```python
df3.head()
```

```python
df4.head()
```

```python
df = pd.concat([df1,df2,df3,df4,df5])
```

```python
df.head()
```

```python
df.tail()
```

```python
df.sample(5)
```

```python
df.sample(5)
```

```python
df.dtypes
```

```python
df["LojaID"] = df["LojaID"].astype("object")
```

```python
df.dtypes
```

```python
#Consultando linhas com valores faltantes
```

```python
df.isnull().sum()
```

```python
#Substituindo os valores nulos pela media
df["Vendas"].fillna(df["Vendas"].mean(),inplace=True)
```

```python
df["Vendas"].mean()
```

```python
df["Vendas"].fillna(0, inplace=True)
```

```python
df.dropna(subset=["Vendas"], inplace=True)
```

```python
df.dropna(how="all", inplace=True)
```

```python
df["Vendas"].mean()
```

```python
#Criando a coluna de receita
df["Receita"] = df["Vendas"].mul(df["Qtde"])
```

```python
df.head()
```

```python
df["Receita/Vendas"] = df["Receita"] / df["Vendas"]
```

```python
df.head()
```

```python
#Retornando a maior receita
df["Receita"].max()
```

```python
#Retornando a menor receita
df["Receita"].min()
```

```python
#nlargest
```

```python
df.nlargest(3,"Receita")
```

```python
#nsmallest
df.nsmallest(3,"Receita")
```

```python
#Agrupamento por cidade
df.groupby("Cidade")["Receita"].sum()
```

```python
#Ordenando o conjunto de dados
df.sort_values("Receita", ascending=False).head(10)
```

```python
#Transformando a coluna de data em tipo inteiro
df["Data"]=df["Data"].astype("int64")
```

```python
df.dtypes
```

```python
df.head()
```

```python
df["Data"] = pd.to_datetime(df["Data"])
```

```python
df.head()
```

```python
df.dtypes
```

```python
#Agrupamento por ano
df.groupby(df["Data"].dt.year)["Receita"].sum()
```

```python
#Criando uma nova coluna com o ano
df["Ano_venda"] = df["Data"].dt.year
```

```python
df.sample(5)
```

```python
#Extraindo o mês e o dia
df["mes_venda"], df["dia_venda"] = (df["Data"].dt.month,df["Data"].dt.day)
```

```python
df.sample(5)
```

```python
#Retornando a data mais antiga
df["Data"].min()
```

```python
#Calculando a diferenca de dias
df["diferenca_dias"] = df["Data"] - df["Data"].min()
```

```python
df.sample(5)
```

```python
#Criando a coluna de trimestre
df["trimestre_venda"] = df["Data"].dt.quarter
```

```python
df.sample(5)
```

```python
df["LojaID"].value_counts(ascending=False)
```

```python
df["LojaID"].value_counts(ascending=False).plot.bar()
```

```python
#Grafico de barras horizontais
df["LojaID"].value_counts(ascending = True).plot.barh();
```

```python
#Grafico de Pizza
df.groupby(df["Data"].dt.year)["Receita"].sum().plot.pie()
```

```python
#Total de vendas por cidade
df["Cidade"].value_counts()
```

```python
#Adicionando um titulo e alterando o nome dos eixos
import matplotlib.pyplot as plt
df["Cidade"].value_counts().plot.bar(title="Total vendas por Cidade")
plt.xlabel("Cidade")
plt.ylabel("Total Vendas")
```

```python
#Alterando a cor
df["Cidade"].value_counts().plot.bar(title="Total vendas por Cidade", color="Red")
plt.xlabel("Cidade")
plt.ylabel("Total Vendas")
```

```python
#Alterando o estilo
plt.style.use("ggplot")
```

```python
df.groupby(df["mes_venda"])["Qtde"].sum().plot(title = "Total de produtos vendidos por mes")
plt.xlabel("Mês")
plt.ylabel("Total Produtos Vendidos");
plt.legend()
```

```python
#Selecionando apenas as vendas de 2019
df_2019 = df[df["Ano_venda"] == 2019]
```

```python
#Total produtos vendisdos por mes
df_2019.groupby(df_2019["mes_venda"])["Qtde"].sum().plot(marker = "o")
plt.xlabel("Mês")
plt.ylabel("Total Produtos Vendidos");
plt.legend();
```

```python
#Histograma
plt.hist(df["Qtde"], color="orangered");
```

```python
plt.scatter(x=df_2019["dia_venda"], y=df_2019["Receita"]);
```

```python
#Salvando em png
df_2019.groupby(df_2019["mes_venda"])["Qtde"].sum().plot(marker = "v")
plt.title("Quantidade de produtos vendidos x mes")
plt.xlabel("Mês")
plt.ylabel("Total Produtos Vendidos");
plt.legend();
plt.savefig("grafico Qtde x Mes.png")
```
